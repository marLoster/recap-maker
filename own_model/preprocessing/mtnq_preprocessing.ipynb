```python
class LemmaMaker:
    
    def __init__(self):
        self.spacy = spacy.load('pl_core_news_lg')
    
    def lemmatise(self, word):
        return self.spacy(word)[0].lemma_
    
lemmatiser = LemmaMaker() 
```

```python
from pyspark.sql import SparkSession

# Step 4: Create a SparkSession
spark = SparkSession.builder \
    .appName("Read CSV with PySpark") \
    .getOrCreate()

# Step 5: Read the CSV file into a DataFrame
# Replace "your_file_path.csv" with the actual path to your CSV file
df = spark.read.csv("mtnq_v1.0.csv", header=True, inferSchema=True)

# Show the DataFrame to check the data
```

```python
def clean_string(string):
    for character in ",.<>:;'\"()*^+-_=":
         string = string.replace(character,"")
    return string
```

```python
def lemmatise_line(line):
    try:
        line = ' '.join(map(lambda x: lemmatiser.lemmatise(x), line.split()))
    except:
        pass
    return line
```

```python
from pyspark.sql.functions import col, udf
from pyspark.sql.types import StringType
clean = udf(lambda x: lemmatise_line(x),StringType()) 
```
