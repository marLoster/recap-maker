```python
import os
import pandas as pd
import numpy as np

from collections import Counter
```

```python
cnt = Counter()
for file in os.listdir('dataframes'):
    data = pd.read_csv(os.path.join('dataframes',file))
    for row in range(len(data)):
        try:
            cnt.update(data.iloc[row]["passage"].lower().split())
            cnt.update(data.iloc[row]["question"].lower().split())
        except:
            pass
```

```python
with open('stop.txt',encoding='utf-8') as stopwords_file:
    stopwords = set(map(lambda x: x[:-1],stopwords_file.readlines()))
```

```python
words = filter(lambda x: x.isalpha(), cnt.keys())
```

```python
cleared_words = filter(lambda x: x not in stopwords and len(x)>2, words)
```

```python
words_counts = map(lambda x: (x, cnt[x]), cleared_words)
```

```python
data100 = filter(lambda x: x[1] >= 100, words_counts)
```

```python
data100 = list(data100)
```

```python
data100 = pd.DataFrame(data100, columns=["word","count"])
data100.to_csv("data100.csv")
```

```python
words = filter(lambda x: x.isalpha(), cnt.keys())
```

```python
cleared_words = filter(lambda x: x not in stopwords and len(x)>2, words)
```

```python
words_counts = map(lambda x: (x, cnt[x]), cleared_words)
```

```python
data10 = filter(lambda x: x[1] >= 10, words_counts)
```

```python
data10 = list(data10)
```

```python
data10 = pd.DataFrame(data10, columns=["word","count"])
data10.to_csv("data10.csv")
```

```python
set10 = set(data10["word"])
set100 = set(data100["word"])

for file in os.listdir('dataframes'):
    with open(os.path.join('data10', file), 'w', encoding='utf-8') as data10file:
        with open(os.path.join('data100', file), 'w', encoding='utf-8') as data100file:
            data = pd.read_csv(os.path.join('dataframes',file))
            data10file.write("question,passage\n")
            data100file.write("question,passage\n")
            
            for row in range(len(data)):
                try:
                    passage = filter(lambda x: x in set10, data.iloc[row]["passage"].lower().split())
                    question = filter(lambda x: x in set10, data.iloc[row]["question"].lower().split())
                    data10file.write(f'{" ".join(passage)},{" ".join(question)}\n')

                    passage = filter(lambda x: x in set100, data.iloc[row]["passage"].lower().split())
                    question = filter(lambda x: x in set100, data.iloc[row]["question"].lower().split())
                    data100file.write(f'{" ".join(passage)},{" ".join(question)}\n')

                except:
                    pass
```

```python
word10array = np.array(data10['word'])
np.save('data10keys.npy', word10array)
```

```python
word100array = np.array(data100['word'])
np.save('data100keys.npy', word100array)
```
